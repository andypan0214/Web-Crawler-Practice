```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import re


new_list = []
url = 'https://money.udn.com/money/index'

## 1.填寫headers
headers = {'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:108.0) Gecko/20100101 Firefox/108.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
#print(response.text)

## 2.抓取即時區域所有新聞title以及新聞內容連結link，並將每篇新聞的連結link append到new_list內
article_title_html = soup.find_all('div',{'class':"story__list story__list--web"}) # find pattern at 最新


#for i in a_tags:
#    print('title: {}, link: {}'.format(i.text, 'https://money.udn.com'+i['href']))
#    new_list.append('https://money.udn.com'+i['href'])

for index, i in enumerate(article_title_html):
    if index <= 3: # get first 4 data sets
        title = i.find('a')["title"].replace(' ','').replace('\n','')
        href_link = "https://money.udn.com"+i.find('a')['href'] #get url link and paste local url at front
        print('title: {}, link: {}'.format(title, href_link))
        new_list.append(href_link)
        
```

```python
from time import sleep
import os
from bs4 import BeautifulSoup
from urllib.request import urlretrieve

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

        
def save_news(url):
    
    ## 3.填寫headers
    headers = {'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:108.0) Gecko/20100101 Firefox/108.0"}

    response = requests.get(url, headers=headers)
    sleep(1)
    
    soup = BeautifulSoup(response.text, 'html.parser')
    ## 4. 使用find_all去抓取每篇新聞的title(指定是哪個tag以及指定id)，將結果所產生之list放到new_title變數內
    new_title = soup.find_all('h1',{'id':"story_art_title"})
    save_path = './news/'+new_title[0].text.replace(':','')
    #save_path = './news/'+new_list[0].text.replace(':','')
    
    article_body = soup.find_all('section',{'id':'article_body'}) #網站改版div -> section
    print('new title: {}'.format(new_title[0].text))
    #print('new title: {}'.format(new_list[0].text))
    
    ## create folder
    createFolder(save_path)
    soup2 = BeautifulSoup(str(article_body[0]), 'html.parser')


    ## save article contens in folder
    with open(save_path+'/contents.txt', 'w', encoding='utf-8') as f:  
        ## 5. 抓取文章內p tags，並將新聞內文寫入contents.txt檔案內
        p_tags = soup2.find_all('p') # find all the p in article 
        print('new content: ')
        for i in p_tags:
            print(i.text)
            f.write(i.text) #store in file unders news directory

    ## save image in article        
    
    ## 6. 抓取文章內所有img tags，並將圖片下載並存入對應新聞資料夾下
    img_tags = soup2.find_all('img')
    print('save image: ')
    for i in img_tags:
        print(i['src'])
        urlretrieve(i['src'], save_path+'/'+i['title']+'.jpg')
    print('==========\n\n')
    

for new_url in new_list:
    print(new_url)
    save_news(new_url)
```
